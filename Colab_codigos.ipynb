**Soma Fibonacci 20 termos//14 linhas** 

```
import numpy as np
lista=np.linspace(0,20)
lista[0]=0
lista[1]=1  
anterior=lista[0]
ultimo=lista[1]
print(lista[0])
print(lista[1])
for n in range(2,20):
  soma=anterior+ultimo
  lista[n]=soma
  anterior=ultimo
  ultimo=soma
  print(lista[n])
```

**Testar se o n° é par**

```
valor = input("Número teste:\n")
valor = int(valor)
while True:
     if valor%2 == 0:
         print("É par!")
         break
     elif valor%2 != 0:
         print("Não é par!")
         break 
```

**Eq. 2° Grau com raízes imaginárias**

```
import math
a = input("Valor de a:\n")
b = input("Valor de b:\n")
c = input("Valor de c:\n")
a=int(a) 
b=int(b)
c=int(c)
while True:
     if a == b == c == 0 :
         print("Ta me tirano né >:(")
         break
     if a == 0:
         if b == 0:
             print("Ah para né chuchu, a resposta é...nada pq só tem valor de",c)
             break    
         if b!= 0:
             x=-c/b
             print("A resposta é",x)
             break
     if a != 0:
         Delta=(b**2)-4*a*c
         print("Delta vale",Delta)
         pergunta = input("Deseja continuar? 0 para Sim, 1 para Não.")
         print("Se o Delta for negativo a sua equação não contém raízes reais.")
         pergunta = int(pergunta)
         if pergunta == 1:
              break
         if pergunta == 0:
              if Delta >= 0:
                   d = math.sqrt(Delta)
                   soma=-b+d              
                   diminui=-b-d
                   x1=soma/(2*a)
                   x2=diminui/(2*a)
                   print("A equação admite raízes reais de valores,",round(x1, 3) , round(x2, 3))
                   break
              if Delta < 0:
                   d = math.sqrt(Delta*-1)
                   b1=-b/(2*a)              
                   b2=d/(2*a)                   
                   print("_____________________________________________")
                   print("As raízes não pertencem ao conjunto dos Reais")
                   print("Valor de x' é",round(b1,3), "+",round(b2, 3),"i")
                   print("Valor de x'' é",round(b1, 3), "-",round(b2, 3),"i")
                   break                      
```
